# Tasks:

1. Convert a color image (RGB) to (HSV/HSI) color space using OpenCV.

2. show the three channels separated.

3. save them.

```python
# 1. Convert a color image (RGB) to (HSV/HSI) color space using OpenCV.
# 2. show the three channels separated.
# 3. save them

from google.colab.patches import cv2_imshow
import cv2

imgrgb = cv2.imread('cat1.png')
cv2_imshow(imgrgb)

# 1. from RGB to HSV
img_hsv = cv2.cvtColor(imgrgb, cv2.COLOR_BGR2HSV)

# type(hsv_img)
cv2_imshow(img_hsv)

# 2. Showing the H, S, and V channels separately

cv2_imshow(img_hsv[:, :, 0]) # H Channel
cv2_imshow(img_hsv[:, :, 1]) # S Channel
cv2_imshow(img_hsv[:, :, 2]) # V Channel

#3.  Save
cv2.imwrite('image_H.png', img_hsv[:, :, 0]) # H Channel

cv2.imwrite('image_S.png', img_hsv[:, :, 1]) # S Channel

cv2.imwrite('image_V.png', img_hsv[: , :, 2]) # V Channel


```

#Tasks:

1. Translate an Image
Write a Python script that loads an image and translates it 100 pixels to the right and 50 pixels down. Then, display the original and translated images side by side.

2. Rotate an Image Without Cropping
Rotate an image by 30 degrees clockwise, ensuring that the entire image is visible and not cropped. Adjust the size of the resulting image window so that the whole rotated image fits within it.

3. Scale an Image
Load an image and resize it, doubling its width while keeping the height the same. Display the original and resized images to compare them.

```python
# 1. Translate an Image Write a Python script that loads an image and translates it 100 pixels to the right and 50 pixels down.
# Then, display the original and translated images side by side.

from google.colab.patches import cv2_imshow
import cv2
import numpy as np
from matplotlib import pyplot as plt


imgT = cv2.imread('cat1.png')

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(imgT, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

num_rows, num_cols = imgT.shape[:2]

translation_matrix = np.float32([ [1,0,100], [0,1,50] ])

img_translation = cv2.warpAffine(imgT, translation_matrix, (num_cols, num_rows) )


plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(img_translation, cv2.COLOR_BGR2RGB))
plt.title('Translation')
```

```python
#2. Rotate an Image Without Cropping Rotate an image by 30 degrees clockwise, ensuring that the entire image is visible and not cropped.
#Adjust the size of the resulting image window so that the whole rotated image fits within it.

rotation_matrix = cv2.getRotationMatrix2D((num_cols/2, num_rows/2), -30, 0.61)
img_rotation = cv2.warpAffine(imgT, rotation_matrix, (num_cols, num_rows))
plt.imshow(cv2.cvtColor(img_rotation, cv2.COLOR_BGR2RGB))
plt.title('Rotation')
```

```python
# 3. Scale an Image Load an image and resize it, doubling its width while keeping the height the same.
#Display the original and resized images to compare them.

# Scale the image by a factor
imgT = cv2.imread('cat1.png')
plt.imshow(cv2.cvtColor(imgT, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.show()

img_scaled = cv2.resize(imgT, None, fx=2, fy=1, interpolation=cv2.INTER_LINEAR)
plt.imshow(cv2.cvtColor(img_scaled, cv2.COLOR_BGR2RGB))
plt.title('Scaling')

```
